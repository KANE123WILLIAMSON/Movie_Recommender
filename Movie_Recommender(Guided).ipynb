```python
import pandas as pd
import numpy as np
```

```python
credits=pd.read_csv("C:/Users/Prasun Das/Downloads/tmdb_5000_credits.csv/tmdb_5000_credits.csv")
movies=pd.read_csv("C:/Users/Prasun Das/Downloads/tmdb_5000_movies.csv/tmdb_5000_movies.csv")
```

```python
credits.head(1)
movies.head(1)
credits.shape
```

```python
# Here we merge both the credits and movies to movies
movies=movies.merge(credits,on="title")
movies.head()
```

```python
# Now we drop some columns by our intuition which shouldnt impact our recommendation system
movies.columns
# genres, id, keywords, overview, popularity, cast_x, crew_x,vote_average,original title
movies=movies[["genres","id","keywords","overview","original_title","popularity","cast","crew","vote_average"]]
# Now lets check if any missing or duplicate values are there
movies.isnull().sum()
# Here 3 values in overview column were missing out of 5500 values so it wont affect the distribution of the data
movies.dropna(inplace=True)
movies.shape
movies.duplicated().sum()
movies.drop_duplicates(inplace=True)
movies.shape
```

```python
# Now as the duplicated and null values are removed
movies.head(1)
# We need to set the formatting of the genres
movies.iloc[0].genres
```

```python
# Lets build a function to rewrite it in a easier format
import ast

def convert1(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i["name"])
  return L
```

```python
movies['genres']=movies.genres.apply(convert1)
movies.head(1)
```

```python
# Now lets convert keywords to a more readable format
movies.iloc[0].keywords
import ast
def convert2(obj):
  L=[]
  for i in ast.literal_eval(obj):
    L.append(i["name"])
  return L
movies['keywords']=movies.keywords.apply(convert2)
movies.head(1)
```

```python
# Now lets come to the cast column
movies.iloc[0].cast
# Now here we are assuming that the names are given in a hierarchial order and we are using only the top 3 names
import ast
def convert3(obj3):
  L=[]
  c=0
  for i in ast.literal_eval(obj3):
    if c!=3:
      L.append(i["name"])
      c+=1
    else:
      break
  return L
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.iloc[0].crew
```

```python
import ast
def convert4(obj):
  L=[]
  for i in ast.literal_eval(obj):
    if i['job']=="Director":
      L.append(i["name"])
      break
  return L
movies['crew']=movies['crew'].apply(convert4)
```

```python
# Extracted the names of the directors only
# Convert the overview column into list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'].apply(lambda x:[i. replace(" ","") for i in x])
movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'].apply(lambda x:[i.replace(" ","") for i in x] )
movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
# Removed the spaces between words
# Lets create a new list called tags
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
movies_new=movies[['id','original_title','tags']]
movies_new
movies_new['tags'] = movies_new['tags'].apply(lambda x: [i.lower() for i in x])
```

```python
movies_new['tags']=movies_new['tags'].apply(lambda x:" ".join(x))
```

```python
# Now we do vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors=cv.fit_transform(movies_new['tags']).toarray()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
# created a function to stem similar words to common words
movies_new['tags']=movies_new['tags'].apply(stem)
```

```python
# Now again if we repeat the vectorization process we get reduced result and accuracy increases
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors1=cv.fit_transform(movies_new['tags']).toarray()
```

```python
pd.DataFrame(cv.get_feature_names_out())
# Now the features with same name get filtered out
# Now we shall be using cosine simmilarity to find out the similar movies
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vectors1)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    if movie not in movies_new['original_title'].values:
        print("Movie not found!")
        return

    movie_index = movies_new[movies_new['original_title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(
        list(enumerate(distance)),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    for i in movies_list:
        print(movies_new.iloc[i[0]].original_title)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pickle

# Compute cosine similarity
similarity = cosine_similarity(vectors1)

# Number of similar movies to store for each movie
TOP_K = 50

top_similar = []

for i in range(len(similarity)):
    # Get indices of the top 50 most similar movies (excluding itself)
    indices = np.argsort(similarity[i])[::-1][1:TOP_K+1]

    # Store (movie_index, similarity_score)
    recommendations = []

    for idx in indices:
        recommendations.append((int(idx), float(similarity[i][idx])))

    top_similar.append(recommendations)

# Save the compressed recommendation list
pickle.dump(top_similar, open("similarity_1.pkl", "wb"))
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pickle

# Compute similarity matrix
similarity = cosine_similarity(vectors1)

TOP_K = 10

top_similar = []

for i in range(len(similarity)):
    # Get indices of the 10 most similar movies (excluding itself)
    indices = np.argsort(similarity[i])[::-1][1:TOP_K+1]

    # Store only the indices
    top_similar.append(indices)

# Save the new similarity file
with open("similarity_2.pkl", "wb") as f:
    pickle.dump(top_similar, f, protocol=pickle.HIGHEST_PROTOCOL)
```
